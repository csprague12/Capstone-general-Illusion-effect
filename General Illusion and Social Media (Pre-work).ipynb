## Cleaning

The <b>purpose</b> of cleaning the data here is to eventually create "helper" columns which will make integrating Twitter's API features easier with our dataset. Since we already know the articles here are "viral," we can include some of Twitter's analytics such as number of likes, retweets, and duration of trending hashtags. 

Start off by removing columns containing the same value because columns without variety are not particularly useful in terms of analysis. 

```python
"This for loop will check for unique values in each column and remove if there is the same value"

for col in viral_clean.columns:
    if len(viral_clean[col].unique()) == 1:
        viral_clean.drop(col,inplace = True,axis = 1)
```

```python
viral_clean.head()
```

```python
viral_clean = viral_clean.reset_index(drop=True)
```

Plan is to make a search phrase based on the article title to eventually feed into the Twitter API's search feature. To save our fingers, it'll be faster to form search words from a unique identifyer rather than the full article title. 

```python
viral_clean['Unique Identifyer'] = pd.Series(range(1,101))
```

```python
viral_clean.head()
```

## Introducing Spacy to Create Key Words

Not going to get into anything super advanced right now - I'm more looking for a "quick and dirty" NLP usage to avoid manually making keywords. 

```python
nlp = spacy.load('en_core_web_sm')
```

```python
"The code below will iterate through the 'Title' column and return keywords for each title"

viral_clean['Keyword'] = viral_clean['Title'].apply(lambda x: list(nlp(x).ents))
```

```python
viral_clean.head(25)
```

<b>Verdict:</b> Might have to implement a more advanced NLP mechanism, but it's a quick way to pick out keywords from an article title.
